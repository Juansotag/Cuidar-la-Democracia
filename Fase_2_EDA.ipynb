# Exploratory Data Analysis (EDA) - Fase 2

Este notebook realiza un análisis exploratorio de los datos de la encuesta 'Cuidar la Democracia', enfocándose en distribuciones univariadas, estadísticos descriptivos y cruces demográficos clave.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de visualización
%matplotlib inline
sns.set_theme(style='whitegrid')
```

```python
# Paleta de colores personalizada
PALETA_DEMOCRACIA = ['#2D245D', '#1BAFD5', '#FFBA27', '#E2E3E3', '#F1EFEF']
sns.set_palette(sns.color_palette(PALETA_DEMOCRACIA))
```

```python
# Cargar datos de la Fase 1 y categorización de municipios
df = pd.read_csv('Fase_1_Limpiado.csv')
df_munis = pd.read_csv('municipios_categorizados.csv')

# Unir categorización
df = df.merge(df_munis, on=['DEPARTAMENTO', 'MUNICIPIO'], how='left')

print(f'Dataset cargado con {df.shape[0]} filas y {df.shape[1]} columnas.')
df.head()
```

```python
def plot_univariate(df, column, title, kind='bar'):
    plt.figure(figsize=(10, 6))
    data = df[column].value_counts(normalize=True) * 100
    
    if kind == 'pie':
        # textprops={'color': 'gray'} pone los porcentajes en gris
        data.plot.pie(autopct='%1.1f%%', colors=PALETA_DEMOCRACIA, startangle=90, textprops={'color': 'gray'})
        plt.ylabel('')
    else:
        sns.barplot(x=data.index, y=data.values, palette=PALETA_DEMOCRACIA)
        # color='gray' y tick_params para ejes en gris
        plt.ylabel('Porcentaje (%)', color='gray')
        plt.tick_params(axis='both', colors='gray')
        plt.xticks(rotation=45)
    
    plt.title(title, fontweight='bold', color='#2D245D')
    plt.tight_layout()
    plt.show()

def plot_histogram(df, column, title, bins=20):
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column].dropna(), bins=bins, color=PALETA_DEMOCRACIA[0], kde=True)
    plt.ylabel('Frecuencia', color='gray')
    plt.xlabel(column, color='gray')
    plt.tick_params(axis='both', colors='gray')
    plt.title(title, fontweight='bold', color='#2D245D')
    plt.tight_layout()
    plt.show()

# --- HISTOGRAMAS ---
cols_hist = ['DURACION', 'ALTITUD', 'HOGARES_MANZANA', 'EDAD']
for col in cols_hist:
    if col in df.columns:
        plot_histogram(df, col, f'Distribución de {col}')

# --- DIAGRAMAS DE BARRAS (Demográficos y otros) ---
cols_bar = ['ESTRATO', 'DEPARTAMENTO', 'PERSONAS_EN_HOGAR', 'SEXO']
for col in cols_bar:
    if col in df.columns:
        plot_univariate(df, col, f'Distribución de {col}')

# --- PREGUNTAS DEL CUESTIONARIO (Barras) ---
# Genera gráficas para P8, P9 y el rango P10 a P15
preguntas_bar = ['P8', 'P9', 'P14']
# Añadir sub-preguntas P10_1...P15_5 dinámicamente
for p in range(10, 16):
    if p != 14: # P14 no suele tener sub-ítems en este dataset
        for sub in range(1, 12): # Ajustar el rango 12 según la cantidad de sub-ítems reales
            preguntas_bar.append(f'P{p}_{sub}')

for preg in preguntas_bar:
    if preg in df.columns:
        plot_univariate(df, preg, f'Pregunta {preg}')

# --- DIAGRAMA DE TORTA (P6) ---
if 'P6_PALABRAS_DEMOCRACIA' in df.columns:
    plot_univariate(df, 'P6_PALABRAS_DEMOCRACIA', 'P6. Palabras sobre Democracia', kind='pie')
    
```

## 1. Distribución General (Análisis Univariado)

Análisis de la proporción de respuestas en variables clave.

```python
def plot_univariate(df, column, title, kind='bar'):
    plt.figure(figsize=(10, 6))
    data = df[column].value_counts(normalize=True) * 100
    
    if kind == 'pie':
        # 1. Agregamos 'textprops' para cambiar el color de los porcentajes dentro de la torta
        data.plot.pie(autopct='%1.1f%%', colors=PALETA_DEMOCRACIA, startangle=90, 
                      textprops={'color': 'gray'}) 
        plt.ylabel('')
    else:
        sns.barplot(x=data.index, y=data.values, palette=PALETA_DEMOCRACIA)
        
        # 2. Cambiamos el color de la etiqueta del eje Y a gris
        # plt.ylabel('Porcentaje (%)', color='gray') 
        
        # 3. 'tick_params' cambia el color de los números (ticks) en ambos ejes
        # plt.tick_params(axis='both', colors='gray') 
        
        plt.xticks(rotation=45)
    
    plt.title(title, fontweight='bold', color='#2D245D')
    plt.tight_layout()
    plt.show()
```

## 2. Estadísticos Descriptivos

Análisis de variables numéricas (P36, P37, etc.).

```python
numeric_cols = [c for c in df.columns if 'NUM_PE' in c or c in ['P37', 'EDAD']]
if numeric_cols:
    stats = df[numeric_cols].describe().T
    stats['Varianza'] = df[numeric_cols].var()
    stats['IQR'] = stats['75%'] - stats['25%']
    display(stats.style.background_gradient(cmap='Blues'))
```

## 3. Correlaciones y Asociaciones (Análisis Bivariado)

Relaciones entre variables categóricas clave.

```python
def plot_bivariate(df, x, y, title):
    plt.figure(figsize=(12, 6))
    ct = pd.crosstab(df[x], df[y], normalize='index') * 100
    ct.plot(kind='bar', stacked=True, color=PALETA_DEMOCRACIA, ax=plt.gca())
    plt.title(title, fontweight='bold')
    plt.ylabel('Porcentaje (%)')
    plt.legend(title=y, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# Ejemplo: Cruce P7 vs Sexo
if 'P7' in df.columns and 'SEXO' in df.columns:
    plot_bivariate(df, 'SEXO', 'P7', 'P7 vs Sexo')
```

## 4. Análisis Cruzado Sensible

Cruces por Capital vs Resto, Categoría de Municipio, Estrato y Actividad.

```python
# Cruce por Capital vs Resto
if 'es_capital' in df.columns and 'P7' in df.columns:
    plot_bivariate(df, 'es_capital', 'P7', 'Satisfacción con la Democracia: Capital vs Resto')
```

```python
# Cruce por Categoría de Municipio
if 'categoria_municipio' in df.columns and 'P7' in df.columns:
    plot_bivariate(df, 'categoria_municipio', 'P7', 'Satisfacción vs Categoría Municipio')
```

## 5. Análisis de Variables de Interés (P37, P23_5)

Exploración de variables críticas para la democracia.

```python
target_vars = ['P37', 'P23_5']
for var in target_vars:
    if var in df.columns:
        plot_univariate(df, var, f'Distribución de {var}')
```
